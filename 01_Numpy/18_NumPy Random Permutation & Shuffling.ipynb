# shuffle() and permutation() 

```python
# Permutation : arrangement of elements like
# [3,2,1] is permutation of [1,2,3] , [2,1,3] and vice versa

# 2 methods

# 1) shuffle()  and 2) permutation()

```

# 1) shuffle()

```python
from numpy import random as rm
import numpy as np

data = np.array([1,2,3,4,5])
rm.shuffle(data)   # it shuffles the original array
print(data)
```

```python
data = np.array([1,2,3,4,5])
rm.permutation(data)  # it not change in orginal array 
print(data)
```

```python
data = np.array([1,2,3,4,5])
print(rm.permutation(data)) # its just printing
print(data)
```
