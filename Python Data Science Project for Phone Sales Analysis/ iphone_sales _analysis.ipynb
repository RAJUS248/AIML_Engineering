```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
data = pd.read_csv('apple_products.csv')
# print(data.head(5))
# print(data.isnull().sum())
# print(data.describe())

highest_rated = data.sort_values(by=["Star Rating"],ascending = False)
# print(highest_rated['Product Name'].head(10))
highest_rated = highest_rated.head(10)
# print(highest_rated)
iphone = highest_rated['Product Name'].value_counts()
print(iphone)
```

```python

# iphone = highest_rated['Product Name'].value_counts()
label = iphone.index
count = highest_rated['Number Of Ratings']
graph = px.bar(highest_rated, x=label, y=count, title="number of ratings of highest rated iphones")

graph.show()
```

```python
# iphone = highest_rated['Product Name'].value_counts()
label = iphone.index
count = highest_rated['Number Of Reviews']
graph = px.bar(highest_rated, x=label, y=count, title="number of Reviews of highest Reviews iphones")

graph.show()
```
