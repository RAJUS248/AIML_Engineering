```python
# Getting some elmt out of an existing array and 
# creating a new array is called filtering
```

```python
import numpy as np
data = np.array([41,42,43,44])
data1 = [True,False,True,False]
final_data = data[data1]
print(final_data)  # false not taken
```

# Create Filter Array

```python
emptydata = []
for num in data:
    if num > 42:
        emptydata.append(True)

    else:
        emptydata.append(False)

new_data = data[emptydata]
print(emptydata)
print(new_data)
```

# even elmnt

```python
data = np.array([1,2,3,4,5,6,7])
emptydata = []

for num in data:
    if num % 2 == 0:
        emptydata.append(True)

    else:
        emptydata.append(False)

new_data = data[emptydata]

print(emptydata)
print(new_data)
```

# alternate mtd

```python
emptydata = data > 2
new_data = data[emptydata]
print(emptydata)
print(new_data)
```

```python
emptydata = data  % 2 == 0
new_data = data[emptydata]
print(emptydata)
print(new_data)
```
