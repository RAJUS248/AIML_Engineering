# adding the Grid lines

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([80,85,90,95,100])
y = np.array([180,185,190,195,200])

font1 = {'family' : 'serif', 'color' : 'blue', 'size' : 10}
font2 = {'family' : 'serif', 'color' : 'red', 'size' : 10}

plt.plot(x,y)
plt.grid()
plt.title("Health Chart", fontdict = font1)
plt.xlabel('avg oxygen', fontdict = font2)
plt.ylabel('calorie', fontdict = font2)
plt.show()
```

# specify the grid line

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([80,85,90,95,100])
y = np.array([180,185,190,195,200])

font1 = {'family' : 'serif', 'color' : 'blue', 'size' : 10}
font2 = {'family' : 'serif', 'color' : 'red', 'size' : 10}

plt.plot(x,y)
plt.grid(axis = 'x',color = 'r',ls = '--', lw = 2)
plt.title("Health Chart", fontdict = font1)
plt.xlabel('avg oxygen', fontdict = font2)
plt.ylabel('calorie', fontdict = font2)
plt.show()
```
