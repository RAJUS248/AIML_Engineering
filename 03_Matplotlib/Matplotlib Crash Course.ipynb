```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# basic graph

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)
plt.title('Our 1st graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20})
plt.xlabel('x Axis' ,fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('y Axis',fontdict={'fontname': 'Comic Sans MS'})
plt.show()
```

```python
x = [0,1,2,3]
y = [0,2,4,6]

plt.figure(figsize = (4,2),dpi=200)
# plt.plot(x,y, label = 'raj' , c = 'red', linewidth = 2, marker = '*', linestyle = '--', markersize=10, markeredgecolor = 'blue')

# or use

plt.plot(x,y, 'rs--', label='raj')


plt.title('Our 1st graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20})
plt.xlabel('x Axis' ,fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('y Axis',fontdict={'fontname': 'Comic Sans MS'})

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

# line number two

```python
x = np.arange(0,4.5,0.5)

plt.figure(figsize = (3,2),dpi=250)  # clarity and resize
plt.plot(x[:5], x[:5]**2 , 'r*--', label = 'x^2')
plt.plot(x[4:], x[4:]**2 , 'y')

plt.legend()
plt.show()
```
