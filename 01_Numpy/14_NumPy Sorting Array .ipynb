# sort()

```python
import numpy as np
arr = np.array([7,5,6,1,3,4,9])
print(np.sort(arr))   # like copy orginal array is no changes
print(arr)
```

# sort by string

```python
string = np.array(['z','g','r','s'])
boolin = np.array(['True','False','False'])
print(np.sort(string))
print(np.sort(boolin))
```

# for 2D array

```python
arr = np.array([[3,5,4],[1,5,3]])
print(np.sort(arr))
```
