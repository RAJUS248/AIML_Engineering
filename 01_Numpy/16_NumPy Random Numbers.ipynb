# rand() randint() choice()

# Random : Not predictable

```python
# import numpy as np
from numpy import random
data = random.randint(100)
print(data)
```

# Float num by rand()

```python
from numpy import random
data = random.rand() # if not passing any to float rand it gives 0 to 1 num in float
print(data)
```

# genrate Random array

```python
data = random.randint(100,size = (5))  # for 1D array
print(data)
```

```python
# for 2D array
data = random.randint(100, size = (3,5))
print(data)
```

```python
# for float 1D and 2D array
data1d = random.rand(5)
data2d = random.rand(3,5)
print(data1d)
print()
print(data2d)
```

# Choice

```python
# 1D array
data = random.choice([3,5,7,9,1,4,6]) # randomly choose the num in this array
print(data)
```

```python
# 2D array
data = random.choice([3,5,7,9,1,4,6],size = (3,5))
print(data)
```
