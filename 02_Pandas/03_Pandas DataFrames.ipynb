# DataFrame : 2D data structure like table rows and column

```python
import pandas as pd

data = {"cal": [420,380,390,500], "time" : [50,60,80,40]}
data2 = pd.DataFrame(data)
print(data2)
```

# locate row

```python
# pandas use loc attribute to return one or more specified row
```

```python
data = {"cal": [420,380,390,500], "time" : [50,60,80,40]}
data2 = pd.DataFrame(data)
print(data2.loc[0])  # 1d
print(data2.loc[[0,1]])  # returning row 0 and 1
```

```python
# named Index: with the index arg, you can name ur own index
```

```python
data = {"cal": [420,380,390,500], "time" : [50,60,80,40]}
data2 = pd.DataFrame(data, index=["d1",'d2','d3','d4'])
print(data2)
print(data2.loc['d1'])   # 1d
print(data2.loc[['d1','d2']]) # 2d
```
