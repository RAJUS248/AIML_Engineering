# Univariate - Numerical

```python
data = {
    'Salary' : [25000,35000,55000,40000,45000,30000,50000,70000,60000,65000]
}
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
df.shape
```

# Line Plot:

```python
plt.plot(df['Salary'], color = 'red')
plt.grid()
plt.show()
```

```python
plt.plot(df['Salary'], color = 'red', marker = 'o')
plt.grid()
plt.show()
```

```python
plt.plot(df['Salary'], color = 'red', marker = 'o', linestyle = '--')
plt.grid()
plt.show()
```

```python
plt.plot(df['Salary'], color = 'red', marker = '*', linestyle = ':',linewidth = '2')
plt.grid()
plt.show()
```

# Histogram

```python
data = {
    'Salary' : [25000,45000,35000,55000,25000,35000,40000,55000,45000,30000,50000,70000,60000,65000]
}

df = pd.DataFrame(data)
plt.hist(df['Salary'])
plt.show()
```

```python
plt.hist(df['Salary'], bins = 5, color = 'green')
plt.show()
```

# Boxplot

```python
plt.boxplot(df['Salary'])
plt.show()
```

```python
df.loc[13] = [0]
plt.boxplot(df['Salary'])
plt.show()
```

```python
df.drop(index = 13, inplace = True)
plt.boxplot(df['Salary'])
plt.show()
```

```python
df.shape
```

# Univariate: Categorical

```python
df['dept'] = ['hr','it','finance','hr','it','finance','hr','it','finance','hr','it','finance']
```

```python
df.head()
```

```python
df.info()
```

# Pie Chart

```python
count = df['dept'].value_counts()
count
```

```python
plt.pie(count)
print(plt.show())
```

```python
plt.pie(count , labels = count.index)
print(plt.show())
```

```python
plt.pie(count , labels = count.index, autopct = '%1.2f')
print(plt.show())
```

```python
plt.pie(count , labels = count.index, autopct = '%1.0f', explode = [0.1,0.1,0])
print(plt.show())
```

```python
plt.pie(count , labels = count.index, autopct = '%1.0f', explode = [0,0.1,0.2])
plt.axis('equal')
print(plt.show())
```

# Countplot

```python
plt.bar(count.index, count)
plt.show()
```

```python
plt.bar(count.index, count, color = ['green','black','red'])
plt.show()
```

# Bivariate - numerical - numerical 

```python
df['age'] = [22,25,29,24,30,35,40,36,28,22,23,25]
```

```python
df.head()
```

# Scatter plot

```python
plt.scatter(df['age'],df['Salary'],color = 'black')
plt.show()
```

# line plot

```python
plt.plot(df['age'],df['Salary'],color = 'black' , marker = "*", linestyle='--')
plt.show()
```

```python
sort_age = df.sort_values('age')
plt.grid()
plt.plot(sort_age['age'],df['Salary'],color = 'black' , marker = "*", linestyle='--')
plt.show()
```

# Bar Chart


```python
plt.bar(sort_age['age'],df['Salary'],color = 'black')
plt.show()
```

```python
plt.bar(sort_age['age'],df['Salary'],color = 'black',align = 'edge')
plt.show()
```
