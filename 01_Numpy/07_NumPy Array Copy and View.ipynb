# diffrence b/w numpy arra copy vs view

# make copy

```python
import numpy as np
arr = np.array([1,2,3,4])
arr2 = arr.copy()
print(arr)
print(arr2)
```

```python
arr = np.array([1,2,3,4,6])
arr2 = arr.copy()
arr[0] = 10
print(arr)
print(arr2)
```

# make view

```python
arr = np.array([1,2,3,4])
arr2 = arr.view()
arr[0] = 10
print(arr)
print(arr2)
```
