```python
# if you change orginal data
# use dropna(inplce = True)
```

```python
data = pd.read_csv('dirtydata.csv') # load and read the file
data.dropna(inplace = True)
print(data.to_string())
```

# replace empty cell into value
# using fillna()

```python
data = pd.read_csv('dirtydata.csv') # load and read the file
data.fillna(130, inplace = True)
print(data.to_string())
```

# to replace only for specific column

```python
data = pd.read_csv('dirtydata.csv') # load and read the file
data['Calories'] = data['Calories'].fillna(130)
print(data.to_string())
```

# Replace the empty cell using mean median mode()

# For Mean()

```python
data = pd.read_csv('dirtydata.csv')
x = data['Calories'].mean()
data['Calories'] = data['Calories'].fillna(x)
print(data.to_string())
```

# For Median()

```python
data = pd.read_csv('dirtydata.csv')
x = data['Calories'].median()
data['Calories'] = data['Calories'].fillna(x)
print(data.to_string())
```

# for Mode()

```python
data = pd.read_csv('dirtydata.csv')
x = data['Calories'].mode()[0]       # for finding one by one num so use [0]
data['Calories'] = data['Calories'].fillna(x)
print(data.to_string())
```
