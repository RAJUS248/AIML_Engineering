# MatPlotLib Line, Line Style, Line Color & Line Width

```python
# linestyle or ls - is used to change the style of the line
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
# plt.figure(figsize = (6,2),dpi=900)
# plt.plot(y, ls = 'dotted')
# plt.plot(y, ls = 'dashed')
plt.plot(y, ls = ':')
plt.show()
```

# line color

```python
plt.plot(y, c = 'r')
plt.show()
```

# line width()

```python
plt.plot(y,lw = '10.5')
plt.show()
```

# ex for multiple line

```python
x = np.array([6,2,7,11])
y = np.array([3,8,1,10])

plt.plot(x,lw = '10')
plt.plot(y)
plt.show()
```
