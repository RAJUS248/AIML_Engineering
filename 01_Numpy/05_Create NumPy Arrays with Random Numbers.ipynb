# Random

# rand()

```python
import numpy as np

var = np.random.rand(4)
print(var)
```

```python
var2 = np.random.rand(2,5)
print(var2)
```

# Randn()

```python
var3 = np.random.randn(6)
print(var3)
```

# Ranf()

```python
var4 = np.random.ranf(5)
print(var4)
```

# Randint()

```python
var5 = np.random.randint(5,20,5)  # (min,max,total_elments)
print(var5)
```
