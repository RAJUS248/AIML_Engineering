# Zeros

```python
import numpy as np

arr_zero = np.zeros(4)
arr_2d_zero = np.zeros((3,4))
print(arr_zero)
print()
print(arr_2d_zero)
```

# Ones

```python
import numpy as np

arr_one = np.ones(4)
arr_2d_one= np.ones((3,4))
print(arr_one)
print()
print(arr_2d_one)
```

# Empty Array


```python
arr_empt = np.empty(4)
print(arr_empt)
```

# Range

```python
ranging = np.arange(5)
print(ranging)
```

# Digonal

```python
arr_dig = np.eye(3)
print(arr_dig)
```

```python
arr_dig = np.eye(3,4)
print(arr_dig)
```

# linspaces

```python
space = np.linspace(1,10,num = 5)
space1 = np.linspace(0,20,num = 5)
print(space)
print(space1)
```
