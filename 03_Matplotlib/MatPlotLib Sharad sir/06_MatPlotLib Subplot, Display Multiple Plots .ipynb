# MatPlotLib Subplot, Display Multiple Plots 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1) # 1 for 1 row  2 fro 2 columns, this plot is the 1 (1,2,1)
plt.plot(x,y)

# plot 2

x = np.array([1,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
# Plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,1,1) # 2 for 2 row  1 fro 1 columns, this plot is the 1 (1,2,1)
plt.plot(x,y)

# plot 2

x = np.array([1,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()
```

```python
# Plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,3,1) 
plt.plot(x,y)
plt.title("Sales", c = 'r')

# plot 2

x = np.array([3,1,2,5])
y = np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title("Sales", c = 'r')

# plot 3

x = np.array([1,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.title("Sales", c = 'r')

# plot 4

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(2,3,4)
plt.plot(x,y)
plt.title("Sales", c = 'r')


# Plot 5
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,3,5) 
plt.plot(x,y)
plt.title("Sales", c = 'r')


# Plot 6
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,3,6) 
plt.plot(x,y)
plt.title("Sales", c = 'r')

plt.suptitle('My Shop')
plt.show()
```
