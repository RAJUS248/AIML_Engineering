# panda series is like a column in a table
# it is 1D array which is holds data of any time

```python
import pandas as pd
arr = [1,7,2]
arr2 = pd.Series(arr)
print(arr2)
```

# labeling - it is used to acccess a specified value

```python
arr = [1,7,2]
arr2 = pd.Series(arr)
print(arr2[1])
```

# with create label you can create your own name labels

```python
arr = [1,7,2]
arr2 = pd.Series(arr, index = ['x','y','z'])
print(arr2)
print(arr2['x'])
```

# also use a key or value like Dictionary

```python
import pandas as pd
cal = {"day1":420, "day2":380, "day3":390}
arr = pd.Series(cal)
print(arr)
```

# access only day1 and day2

```python
cal = {"day1":420, "day2":380, "day3":390}
result = pd.Series(cal, index=["day1","day2"])
print(result)
```

# Dataframe

```python
# dataframe: data sets in pandas are useually multidimentional tables 
# and they are called dataframes(whole table)
```

```python
arr = {"cal": [420,380,390], "duration": [50,40,45]}
arr2 = pd.DataFrame(arr)
print(arr2)
```
