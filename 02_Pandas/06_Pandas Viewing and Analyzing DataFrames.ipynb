# viewing the Data using head() method

```python
# head(10) shows the 1st 10 rows
```

```python
import pandas as pd

data = pd.read_csv('data.csv')
print(data.head(10))
```

# viewing the Data using tail() method

```python
# tail(10) shows the 1st 10 rows
```

```python
data = pd.read_csv('data.csv')
print(data.tail(10))
```

# if u want info about the data in the dataframe : via info()
