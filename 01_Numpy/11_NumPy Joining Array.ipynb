# joining the numpy array : concatinate

```python
import numpy as np
arr = np.array([1,2,3])
arr2 = np.array([4,5,7])
arr3 = np.concatenate((arr,arr2))
print(arr3)
```

# joining 2D array

```python
arr = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
arr3 = np.concatenate((arr,arr2),axis = 1)  # axis is means row means 0th is for 0th only
print(arr3)
```

# joining using stack

```python
arr = np.array([1,2,3])
arr2 = np.array([4,5,7])
arr3 = np.stack((arr,arr2), axis = 1)
print(arr3)
```

# stacking along with rows

```python
arr = np.array([1,2,3])
arr2 = np.array([4,5,7])
arr3 = np.hstack((arr,arr2))
print(arr3)
```

# stacking along with columns

```python
arr = np.array([1,2,3])
arr2 = np.array([4,5,7])
arr3 = np.vstack((arr,arr2))
print(arr3)
```

# staking along with hight

```python
arr = np.array([1,2,3])
arr2 = np.array([4,5,7])
arr3 = np.dstack((arr,arr2))
print(arr3)
```
