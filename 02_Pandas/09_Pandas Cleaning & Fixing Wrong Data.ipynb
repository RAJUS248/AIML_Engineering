```python
# for larger data set: we make some rules like for in duration we apply the rule like 
# the value is more then 0 and less then or equal to 60
```

```python
data = pd.read_csv('dirtydata.csv')

for i in data.index:
    if data.loc[i, "Duration"] > 120:
        data.loc[i, "Duration"] = 120

print(data.to_string())
        
```

# you can also remove the data in large data set

```python
data = pd.read_csv('dirtydata.csv')

for i in data.index:
    if data.loc[i, "Duration"] > 120:
        data.drop(i, inplace=True)

print(data.to_string())
```
