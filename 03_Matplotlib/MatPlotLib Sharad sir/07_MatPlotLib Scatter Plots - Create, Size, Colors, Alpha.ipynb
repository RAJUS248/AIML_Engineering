# MatPlotLib Scatter Plots - Create, Size, Colors, Alpha

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

# compair 2 plots

```python
# Day 1 

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])  # age
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) # speed

plt.scatter(x,y, c = 'r')

# day 2

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])  # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85]) # speed

plt.scatter(x,y, c='k')
plt.show()
```

# also change the color of each dot

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])  # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,9]) # speed

colors = ('red','green','blue','black','pink','yellow','orange','purple','gray','cyan','magenta','brown','beige')
plt.scatter(x,y,c = colors)
plt.show()
```

# color map

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])  # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,9]) # speed

colors = np.array([0,13,22,32,46,51,68,75,89,59,103,113,123])

# plt.scatter(x,y,c = colors, cmap = 'viridis')
# or
plt.scatter(x,y,c = colors)
plt.show()
```

# color bar

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])  # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,9]) # speed

colors = np.array([0,13,22,32,46,51,68,75,89,59,103,113,123])

plt.scatter(x,y,c = colors)
plt.colorbar()
plt.show()
```

# Change the Size 

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])  # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,9]) # speed

sizes = np.array([0,13,22,32,46,51,68,75,89,59,103,113,123])

plt.scatter(x,y,s = sizes)
plt.show()
```

# Alpha() : transferency

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])  # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,9]) # speed

sizes = np.array([0,13,22,32,46,51,68,75,89,59,103,113,123])

plt.scatter(x,y,s = sizes, alpha = 0.6)
plt.show()
```

# Combine color size and Alpha

```python

x = np.random.randint(100, size = (100)) # age
y = np.random.randint(100, size = (100)) # speed

sizes = 10 * np.random.randint(100, size = (100))
colors = np.random.randint(100, size = (100))

plt.scatter(x,y, s = sizes, alpha = 0.6 , c = colors, cmap = 'nipy_spectral') # cmap renbow type for me
plt.colorbar()
plt.show()
```
