# MatPlotLib Labels and Title

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y)

plt.title('Health Monitor')
plt.xlabel('avrage oxygen')
plt.ylabel('calorie')

plt.show()
```

# fontdict

```python
fonts ={'family':'serif', 'color':'blue', 'size':20} 
fonts2 ={'family':'serif', 'color':'red', 'size':15} 
plt.plot(x,y)

plt.title('Health Monitor',fontdict = fonts )
plt.xlabel('avrage oxygen', fontdict = fonts2)
plt.ylabel('calorie',fontdict = fonts2)

plt.show()
```

```python
fonts ={'family':'serif', 'color':'blue', 'size':20} 
fonts2 ={'family':'serif', 'color':'red', 'size':15} 
plt.plot(x,y)

plt.title('Health Monitor',fontdict = fonts, loc = 'left' )
plt.xlabel('avrage oxygen', fontdict = fonts2)
plt.ylabel('calorie',fontdict = fonts2)

plt.show()
```
